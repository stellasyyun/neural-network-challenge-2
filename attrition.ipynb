## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition','Department']]
```

```python
# Create a list of at least 10 column names to use as X data
X = ['Education','Age','DistanceFromHome','JobSatisfaction','OverTime','StockOptionLevel','WorkLifeBalance','YearsAtCompany','YearsSinceLastPromotion','NumCompaniesWorked']

# Create X_df using your selected columns
X_df = attrition_df[X]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train.loc[:, 'OverTime'] = X_train['OverTime'].replace({'Yes': 1, 'No': 0})
X_test.loc[:, 'OverTime'] = X_test['OverTime'].replace({'Yes': 1, 'No': 0})
X_train['OverTime'].value_counts()
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_train = scaler.fit_transform(X_train)

# Scale the training and testing data
X_test = scaler.transform(X_test)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
encoder.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_department = encoder.fit_transform(y_train[['Department']])
y_test_department = encoder.transform(y_test[['Department']])
y_train_department
```

```python
# Create a OneHotEncoder for the Attrition column
att_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
att_encoder.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_att = att_encoder.fit_transform(y_train[['Attrition']])
y_test_att = att_encoder.transform(y_test[['Attrition']])
y_train_att
```

## Part 2: Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data.
num = X_train.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(num,), name='input')

# Create at least two shared layers
shared1 = layers.Dense(64, activation='relu', name='shared1')(input_layer)
shared2 = layers.Dense(128, activation='relu', name='shared2')(shared1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(32, activation='relu', name='department_hidden')(shared2)

# Create the output layer
department_output = layers.Dense(3, activation='sigmoid', name='department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(32, activation='relu', name='attrition_hidden')(shared2)

# Create the output layer
attrition_output = layers.Dense(2, activation='sigmoid', name='attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[attrition_output, department_output])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics={'attrition_output': ['accuracy'], 'department_output': ['accuracy']}
)

# Summarize the model
model.summary()
```

```python
# Train the model
history = model.fit(
    X_train,
    {'attrition_output': y_train_att, 'department_output': y_train_department},
    epochs=100,
    batch_size=32,
    validation_data=(X_test, {'attrition_output': y_test_att, 'department_output': y_test_department}),
    verbose=1
)
```

```python
# Evaluate the model with the testing data
results = model.evaluate(X_test, {'attrition_output': y_test_att, 'department_output': y_test_department})
```

```python
# Print the accuracy for both department and attrition
attrition_accuracy = results[3]
department_accuracy = results[4]
print(f"Attrition predictions accuracy: {attrition_accuracy}")
print(f"Department predictions accuracy: {department_accuracy}")
```
